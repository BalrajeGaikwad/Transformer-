```python
from transformers import pipeline
classifier=pipeline('sentiment-analysis')
```

```python
classifier('we are very happy to show you the transformers library')
```

```python
classifier('my mood was not good')
```

```python
classifier('i dont know what is going in my life')
```

```python
results=classifier(['we are very happy to show you the transformers library','my mood was not good','i dont know what is going in my life'])

for result in results:
  print(f"label:{result['label']}, with score: {round(result['score'],4)}")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

model_name = "NousResearch/Minos-v1"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

output = generator("What is the capital of France?", max_new_tokens=50)
print(output[0]["generated_text"])
```
